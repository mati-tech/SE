Regression is a statistical technique used to analyze the relationship between a dependent variable (output variable) and one or more independent variables (input variables). The goal of regression is to find a mathematical model that can predict the value of the dependent variable based on the values of the independent variables. The output of a regression analysis is a regression equation that can be used to predict the value of the dependent variable for any given values of the independent variables. Regression is commonly used in various fields such as finance, economics, social sciences, and engineering to make predictions or understand the relationship between variables.

Regression analysis, is a statistical method used to study the relationship between a dependent variable and one or more independent variables. It explains that regression analysis involves developing an equation that describes how the dependent variable changes as the independent variable(s) change, and that this equation can be used to make predictions about the dependent variable. The passage also describes different types of regression models, such as linear, exponential, and hyperbolic regression, and explains that the goal of regression analysis is to determine the best model to fit the data and estimate the unknown parameters in the equation.

Linear Regression
Linear regression is a model that describes the relationship between a dependent variable x and one or more independent variables (factors, regressors) with a linear function. Linear regression is used to determine the "line of best fit" through a set of data points and has become a simple predecessor to nonlinear methods that are used to train neural networks.

The goal of linear regression is to find the line that best fits these points. Recall that the general equation for a straight line is ùëì(ùë•) = ùëè + ùëö ‚ãÖ ùë•, where ùëö is the slope of the line and ùëè is its intercept.

Loss function - mean squared error
The loss function is a measure of how many errors our linear regression makes on a dataset. Although there are different loss functions, they all calculate the distance between the predicted value ùë¶(ùë•) and its actual value.

One very common loss function is mean squared error (MSE). To calculate MSE, we simply take all the error values, square their lengths, and average them.

Extrapolation problem
Suppose we have many experimental data points. We need to draw a curve through them that passes as close as possible to these points. At the same time, it is necessary to minimize the mean squared error (MSE).

To solve this problem, there are many libraries in Python. The most common ones are:

numpy - numpy.linalg.lstsq
scipy - scipy.linalg (contains all the functions from numpy.linalg plus some new functions that are not in numpy.linalg).

1-1-1 example: 
–ü—Ä–æ–≤–µ–¥–µ–º –ø—Ä—è–º—É—é ùë¶ = ùëöùë• + ùëè —á–µ—Ä–µ–∑ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∞–ª—å–Ω—ã–µ —Ç–æ—á–∫–∏

```python
import numpy as np 
import matplotlib.pyplot as plt 

x = np.array([0,1,2,3])
y = np.array([-1,0.2,0.9,2.1])

A = np.vstack([x,np.ones(len(x))]).T #it makes an array of x and len of x  and T makes the transpose 
print(A)

m, c = np.linalg.lstsq(A,y, rcond=None)[0] #this will find the coefficient for the m and c to fit the data points 

print ('m:', m,'c:', c)

plt.plot(x,y,'o', label = 'Given data', markersize = 12) 
plt.plot (x,m*x +c , 'r', label= 'linear extrapolation', markersize = 10)
plt.legend() #for drawing some labels or something 
plt.show()
```

```python
import matplotlib.pyplot as plt
import scipy as sp
from numpy import *
from numpy.random import *
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

# Define the function to fit
def f(x, b0, b1, b2):
    return b0 + b1 * np.exp(-b2 * (x**2))

# Generate some data points with noise
xdata = np.linspace(0, 5, 50)
beta = (0.25, 0.75, 0.5) # Initial guess for the parameters
y = f(xdata, *beta)
ydata = y + 0.05 * np.random.rand(len(xdata))

# Fit the function to the data using curve_fit
beta_opt, beta_cov = sp.optimize.curve_fit(f, xdata, ydata)

# Compute the linear deviation and the sum of the squared residuals
lin_dev = sum(beta_cov[0])
residuals = ydata - f(xdata, *beta_opt)
fres = sum(residuals**2)

# Plot the results
fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2, label='Initial guess')
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2, label='Fitted curve')
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x,\beta)$", fontsize=18)
ax.legend(fontsize=14)

plt.show()

print(xdata)
print(ydata)



# The code generates some data points with noise, fits 
# a function to the data using curve_fit from the scipy.optimize
# module, computes the linear deviation and the sum of the 
# squared residuals, and plots the results. The function being
# fitted is defined by the f function, which takes three
# parameters (b0, b1, b2) and returns a value for each x value
# passed to it. The initial guess for the parameters is given
# by the beta tuple, which is used as the starting point for
# the optimization algorithm used by curve_fit. Finally, the 
# xdata and ydata arrays are printed to the console.
```

```python
import matplotlib.pyplot as plt
import scipy as sp
from numpy import *
from numpy.random import *
import matplotlib.pyplot as plt
import numpy as np
from  scipy.optimize import  curve_fit

# _____________________________1______________________________
beta = (0.25,0.75)

def f(x,b0,b1):
    return b0 + b1*x

xdata = np.linspace(0,5,50)

y = f(xdata, *beta)

ydata = y + 0.05 * np.random.rand(len(xdata))

beta_opt, beta_cov = sp.optimize.curve_fit(f,xdata,ydata)


# beta_opt is a NumPy array that contains the optimized values of the parameters (coefficients) b0 and b1 that minimize the sum of squares of the residuals between the observed data ydata and the fitted curve f(xdata, *beta_opt).

# beta_cov is the estimated covariance matrix of beta_opt. 
# It provides a measure of the uncertainty of the estimated 
# parameters. The diagonal elements of beta_cov represent the
# variance of the corresponding parameter estimates, while the
# off-diagonal elements represent the covariances between the
# parameter estimates.

lin_dev = sum(beta_cov[0])
print(beta_opt)

residuals = ydata -f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

# lin_dev = sum(beta_cov[0]) calculates the linear deviation,
# which is the sum of the diagonal elements of the covariance
# matrix beta_cov obtained from the curve_fit function. The diagonal elements of this matrix represent the variances of the estimated parameters beta_opt.

# residuals = ydata -f(xdata, *beta_opt) computes the 
# residuals, which are the differences between the observed 
# ydata values and the values predicted by the fitted curve 
# evaluated at xdata using the estimated parameters beta_opt.

# fres = sum(residuals**2) calculates the sum of squared
# residuals, which is a measure of the goodness of fit of 
# the curve. A smaller value of the sum of squared residuals
# indicates a better fit.



fig,ax = plt.subplots()
ax.scatter(xdata,ydata)

ax.plot(xdata,y,'r',lw =2 )
ax.plot(xdata,f(xdata,*beta_opt),'b',lw = 2)

ax.set_xlim(0,5)
ax.set_xlabel(r"$x$", fontsize = 18)
ax.set_ylabel(r"$f(x,\beta)$", fontsize = 18)
plt.title('x')
plt.show()

# _____________________________2______________________________

# Define the function f(x,b0,b1,b2) with three parameters
def f(x, b0, b1, b2):
    return b0 + b1 * x + b2 * x**2

# Define the true parameters beta and generate data from f(x,beta)
beta = (0.25, 0.75, 0.5)
xdata = np.linspace(0, 5, 50)
y = f(xdata, *beta)

# Add some random noise to the data
ydata = y + 0.05 * np.random.rand(len(xdata))

# Use curve_fit to find the optimal parameters beta_opt and the covariance matrix beta_cov
beta_opt, beta_cov = curve_fit(f, xdata, ydata)

# Calculate the sum of the diagonal elements of the covariance matrix to get the linear deviation
lin_dev = np.sum(beta_cov.diagonal())

# Print the optimal parameters and the linear deviation
print("Optimal parameters:", beta_opt)
print("Linear deviation:", lin_dev)

# Calculate the residuals and the sum of squared residuals
residuals = ydata - f(xdata, *beta_opt)
fres = np.sum(residuals**2)

# Print the sum of squared residuals
print("Sum of squared residuals:", fres)

# Plot the data points and the true function f(x,beta) in red
# Plot the fitted function f(x,beta_opt) in blue
fig, ax = plt.subplots()
ax.scatter(xdata, ydata)
ax.plot(xdata, y, 'r', lw=2)
ax.plot(xdata, f(xdata, *beta_opt), 'b', lw=2)
ax.set_xlim(0, 5)
ax.set_xlabel(r"$x$", fontsize=18)
ax.set_ylabel(r"$f(x,\beta)$", fontsize=18)
plt.title('x**2')
plt.show()


# _____________________________3______________________________


# Define the true function parameters and the function
beta = (1,2)
def f(x,b0,b1):
    return b0 + b1*np.log(x)

# Generate some noisy data
xdata = np.linspace(1,5,50)
y = f(xdata, *beta)
ydata = y + 0.05 * np.random.rand(len(xdata))

# Fit the curve using scipy.optimize.curve_fit
beta_opt, beta_cov = sp.optimize.curve_fit(f,xdata,ydata)

# Calculate the linear deviation
lin_dev = sum(beta_cov[0])

# Print the optimized parameters and the sum of the residuals squared
print(beta_opt)
residuals = ydata -f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

# Create a plot of the data and the fitted curve
fig,ax = plt.subplots()
ax.scatter(xdata,ydata)
ax.plot(xdata,y,'r',lw =2 )
ax.plot(xdata,f(xdata,*beta_opt),'b',lw = 2)
ax.set_xlim(0,5)
ax.set_xlabel(r"$x$", fontsize = 18)
ax.set_ylabel(r"$f(x,\beta)$", fontsize = 18)
plt.title('ln(x)')
plt.show()


# _____________________________4______________________________

# Define initial parameters
beta = (1,2)

# Define the function we want to fit to the data
def f(x,b0,b1):
    return b0 * x **b1

# Generate some test data for fitting
xdata = np.linspace(1,5,50)
y = f(xdata, *beta)
ydata = y + 0.05 * np.random.rand(len(xdata))

# Fit the function to the data
beta_opt, beta_cov = sp.optimize.curve_fit(f,xdata,ydata)

# Calculate the sum of the diagonal elements of the covariance matrix
# to estimate the linear deviation of the fit
lin_dev = sum(beta_cov[0])

# Print the optimized parameters and the residual sum of squares
print(beta_opt)
residuals = ydata -f(xdata, *beta_opt)
fres = sum(residuals**2)
print(fres)

# Plot the data and the fit
fig,ax = plt.subplots()
ax.scatter(xdata,ydata)
ax.plot(xdata,y,'r',lw =2 )
ax.plot(xdata,f(xdata,*beta_opt),'b',lw = 2)
ax.set_xlim(0,5)
ax.set_xlabel(r"$x$", fontsize = 18)
ax.set_ylabel(r"$f(x,\beta)$", fontsize = 18)
plt.title('x**b1')
plt.show()


# The two printed values before the graph are:

# beta_opt: This is a NumPy array containing the optimal values of the parameters b0 and b1 that minimize the sum of the squared residuals between the predicted and actual y values. These values are obtained by fitting the function f(x,b0,b1) to the data using the curve_fit function from the scipy.optimize module.
# 
# fres: This is a scalar value representing the sum of the squared residuals between the predicted and actual y values. It is computed using the optimal values of b0 and b1 obtained from the curve_fit function. This value is an indicator of how well the model fits the data, with lower values indicating a better fit.
```

–∏–∑–≤–µ—Å—Ç–Ω—ã–π –º–µ—Ç–æ–¥ numpy.linalg.lstsq.


```python
import numpy as np

y = [1,2,3,4,3,4,5,3,5,5,4,5,4,5,4,5,6,0,6,3,1,3,1]
X = [[0,2,4,1,5,4,5,9,9,9,3,7,8,8,6,6,5,5,5,6,6,5,5],
     [4,1,2,3,4,5,6,7,5,8,7,8,7,8,7,8,6,8,9,2,1,5,6],
     [4,1,2,5,6,7,8,9,7,8,7,8,4,4,3,1,2,3,4,1,3,9,7]]


X =np.transpose(X)
X = np.c_[X,np.ones(X.shape[0])]
linreg = np.linalg.lstsq(X,y,rcond=None)[0]

# multiple regression means that we are modeling the relationship between a dependent variable and multiple independent variables, using a linear equation. It is called "multiple" because there are multiple independent variables, not because we have multiple regression lines. The regression equation in multiple regression has the form:

# y = b0 + b1x1 + b2x2 + ... + bn*xn

# where y is the dependent variable, x1, x2, ..., xn are the 
# independent variables, and b0, b1, b2, ..., bn are the 
# coefficients of the equation. The goal of multiple regression 
# is to find the values of the coefficients that best fit the 
# data, so we can use the equation to make predictions about
# the dependent variable based on the independent variables.
# b0 + b1x1 + b2x2 + ... + bn*xn
print(linreg)
```

```python
import matplotlib.pyplot as plt
import seaborn as seabornInstance

from  sklearn.model_selection import train_test_split
from  sklearn.linear_model import LinearRegression
from  sklearn import metrics

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas import DataFrame,Series
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# In this code, we are performing linear regression on a dataset with 
# three independent variables (x1, x2, x3) and one dependent variable (y).

# First, we create two dataframes (df1 and df2) from the lists of x 
# and y values. We then concatenate these dataframes into a single dataset 
# using the pandas concat() function, which we assign to the variable 
# 'dataset'.
# 
# We then split the dataset into training and testing sets using the 
# train_test_split() function from scikit-learn. We assign 20% of the data 
# to the testing set and the remaining 80% to the training set.

# Next, we create a LinearRegression object and fit it to the training
# data using the fit() method. We then use the coefficients of the regression to create a dataframe that shows the impact of each independent variable on the dependent variable.

# Finally, we make predictions using the test data and calculate the 
# mean squared error between the predicted values and the actual values 
# using the mean_squared_error() function from scikit-learn.

y = [1,2,3,4,3,4,5,3,5,5,4,5,4,5,4,5,6,0,6,3,1,3,1]
X = [[0,2,4,1,5,4,5,9,9,9,3,7,8,8,6,6,5,5,5,6,6,5,5],
     [4,1,2,3,4,5,6,7,5,8,7,8,7,8,7,8,6,8,9,2,1,5,6],
     [4,1,2,5,6,7,8,9,7,8,7,8,4,4,3,1,2,3,4,1,3,9,7]]


new_y = np.array(y)
new_y = new_y.transpose()
df1 = pd.DataFrame(new_y)


new_X = np.array(X)
new_X = new_X.transpose()

df2 = pd.DataFrame(new_X)

df1 = df1.rename(columns={0:'y'}, inplace= False)
df2 = df2.rename(columns={0:'x1',1:'x2',2:'x3'},inplace= False)

frames = [df1,df2]


dataset = pd.concat([df1,df2], axis=1,join="inner")
dataset.head()


print(dataset.head())

print(dataset.shape)
dataset.describe()

X= dataset[['x1','x2','x3']]
y = dataset['y']

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.2,random_state=0)

regressor = LinearRegression()
regressor.fit(X_train,y_train)

coeff_df =  pd.DataFrame(regressor.coef_,X.columns, columns=["Coefficient"])
print(coeff_df)

y_pred = regressor.predict(X_test)

df = pd.DataFrame({"Actual":y_test,'Predicted':y_pred})
print(df)

print("Mean Squared Error:", metrics.mean_squared_error(y_test,y_pred))
```

–∫–∞—á–µ—Å—Ç–≤–µ –∫—Ä–∞—Å–Ω–æ–≥–æ –≤–∏–Ω–∞ (1) –∏ (2) —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ. –ù–∞–π–¥–∏—Ç–µ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç—ã 
–º–Ω–æ–∂–µ—Å—Ç–≤–µ–Ω–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏. –ü–æ—Å—Ç—Ä–æ–π—Ç–µ –ø—Ä–æ–≥–Ω–æ–∑.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# In this code, you are reading two different CSV files from GitHub
# URLs using Pandas read_csv function. Then, you are splitting the 
# datasets into training and testing sets using train_test_split 
# function from scikit-learn library. After that, you are fitting 
# separate linear regression models using the training sets for each 
# dataset and making predictions on the testing sets using the predict
# method of each model. Finally, you are calculating and printing the mean 
# squared error for
# each model using the mean_squared_error function from scikit-learn.

url1 = 'https://raw.githubusercontent.com/likarajo/petrol_consumption/master/data/petrol_consumption.csv'
url2 = 'https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv'

dataset1 = pd.read_csv(url1)
dataset2 = pd.read_csv(url2)

X_train1, X_test1, y_train1, y_test1 = train_test_split(dataset1.iloc[:, :-1],
                                                        dataset1.iloc[:, -1],
                                                        test_size=0.8,
                                                        random_state=0)

X_train2, X_test2, y_train2, y_test2 = train_test_split(dataset2.iloc[:, :-1],
                                                        dataset2.iloc[:, -1],
                                                        test_size=0.8,
                                                        random_state=0)

regressor1 = LinearRegression()
regressor1.fit(X_train1, y_train1)

regressor2 = LinearRegression()
regressor2.fit(X_train2, y_train2)

y_pred1 = regressor1.predict(X_test1)
df1 = pd.DataFrame({'Actual': y_test1, 'Predicted': y_pred1})
display(df1)

y_pred2 = regressor2.predict(X_test2)
df2 = pd.DataFrame({'Actual': y_test2, 'Predicted': y_pred2})
display(df2)

print('Mean Squared Error first csv: ', metrics.mean_squared_error(y_test1, y_pred1))
print('Mean Squared Error second csv: ', metrics.mean_squared_error(y_test2, y_pred2))
# The mean squared error (MSE) indicates the average 
# squared difference between the predicted and actual values.
# A high MSE value indicates that the model is not able to accurately 
# predict the outcome, whereas a low MSE value indicates
# a good fit of the model to the data.
```

–º–µ—Ç–æ–¥–∞ –Ω–∞–∏–º–µ–Ω—å—à–∏—Ö –∫–≤–∞–¥—Ä–∞—Ç–æ–≤. –†–µ–∑—É–ª—å—Ç–∞—Ç—ã –≤—ã–≤–æ–¥—è—Ç—Å—è –≤ –≤–∏–¥–µ —Ç–∞–±–ª–∏—Ü

```python
import numpy as np
import matplotlib.pyplot as plt

# Variant 
x = np.array([4.0, 4.2, 4.4, 4.6, 4.8, 5.0])
y = np.array([4.0, 3.0, 6.0, 6.0, 4.0, 4.0])

A = np.vstack([x, np.ones(len(x))]).T
print(A)

m, c = np.linalg.lstsq(A, y, rcond = None)[0]
print(m, c)

plt.plot(x, y, 'o', label = '–ò—Å—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ', markersize = 10)
plt.plot(x, m*x + c, 'r', label = '–õ–∏–Ω–µ–π–Ω–∞—è —ç–∫—Å—Ç—Ä–∞–ø–æ–ª—è—Ü–∏—è')
plt.legend()
plt.show()

m = np.vstack((x ** 2, x, np.ones(6))).T

s = np.linalg.lstsq(m, y, rcond = None)[0]

x_prec = np.linspace(4, 5, 6)

plt.plot(x, y, 'D')
plt.plot(x_prec, s[0] * x_prec ** 2 + s[1] * x_prec + s[2], '-', lw = 2)
plt.grid()
print(s)
```
